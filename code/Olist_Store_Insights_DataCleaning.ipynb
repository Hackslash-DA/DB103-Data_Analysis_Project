## Dropping null vaues

```python
df3 = df2.dropna(axis = 0)
df2.reset_index(drop=True,inplace=True)
```

## Dropping columns that has the least correlation with the target variable

```python
df3.drop(labels=['product_length_cm','product_height_cm','product_width_cm'], axis=1)
```

```python
df3.drop(labels='product_name_lenght', axis=1)
```

## Renaming Columns

```python
df4=df3.rename(columns={"olist_orders_dataset.order_status": "order_status",
                    "olist_orders_dataset.order_purchase_timestamp": "order_purchase_Time",
                       "olist_orders_dataset.order_approved_at":"order_approved_at",
                       "olist_orders_dataset.order_delivered_carrier_date":"order_delivered_carrier_date",
                       "olist_orders_dataset.order_delivered_customer_date":"order_delivered_customer_date",
                       "olist_orders_dataset.order_estimated_delivery_date":"order_estimated_delivery_date"})
```

```python
df4.info()
```

## Creating a custom Column called "Delivery delay"

```python
df4["delivery_delay"]= (df4["order_estimated_delivery_date"]-df4["order_delivered_customer_date"]).astype('timedelta64[D]')
```

```python
df4.head()
```

```python
df4["order_estimated_delivery_date"]
```

```python
df4["order_delivered_customer_date"]
```

## Creating a custom Column called "Days to deliver"

```python
df4["Days_to_Deliver"]= (df4["order_delivered_customer_date"]-df4["order_approved_at"]).astype('timedelta64[D]')
```

```python
df4.head()
```

## Changing the datatype of newly created columns

```python
df4["Days_to_Deliver"]=df4["Days_to_Deliver"].astype(int)
df4["delivery_delay"]=df4["delivery_delay"].astype(int)
```

```python
df4.head()
```

## Creating a cloumn called "Delayed"

This column has information on delayed orders. 
This is a categorical column that has only two values called “Delayed” and “Not delayed”

```python
df4['delayed'] = np.where((df4['order_delivered_customer_date'] > df4['order_estimated_delivery_date']),"delayed","Not delayed")
                     
```

```python
df4.head()
```
